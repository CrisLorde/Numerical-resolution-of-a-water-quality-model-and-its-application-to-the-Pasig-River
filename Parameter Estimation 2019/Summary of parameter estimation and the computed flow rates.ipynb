## Summary of computed flow rates

```python
import matplotlib.pyplot as plt 
import numpy as np 
  
# create data 
x = np.arange(3) 
Jan = [25.03,3.09,193.41] 
Feb = [250.20,15.05,102.13] 
Mar = [98.67,27.09,22.24]
Apr = [2.88,7.08,87.86]  
May = [67.86, 2.74,12.75]
Jun = [146.37,7.90,154.28] 
Jul = [185.23,7.23,10.20] 
Aug = [13.52,133.61,141.75] 
Sep = [136.18,27.36,178.46] 
Oct = [184.28,71.87,186.81] 
Nov = [196.20,30.86,174.73]

width = 0.08

# plot data in grouped manner of bar type 
plt.bar(x-0.40, Jan, width, color='red', label = "Jan") 
plt.bar(x-0.32, Feb, width, color='#f23000', label = "Feb") 
plt.bar(x-0.24, Mar, width, color='#e65c00', label = "Mar")
plt.bar(x-0.16, Apr, width, color='#d98200', label = "Apr") 
plt.bar(x-0.08, May, width, color='#cca300', label = "May") 
plt.bar(x, Jun, width, color='#bfbf00', label = "Jun")
plt.bar(x+0.08, Jul, width, color='#8fb300', label = "Jul") 
plt.bar(x+0.16, Aug, width, color='#64a600', label = "Aug") 
plt.bar(x+0.24, Sep, width, color='#3d9900', label = "Sep") 
plt.bar(x+0.32, Oct, width, color='#1c8d00', label = "Oct") 
plt.bar(x+0.40, Nov, width, color='green', label = "Nov") 
plt.axhline(y=0, color='black', linestyle='-')


plt.xticks(x, ['Bambang', 'Guadalupe', 'Lambingan']) 
plt.xlabel("Stations") 
plt.ylabel("flow rate $Q$") 
#plt.ylim(-1, 1)
plt.legend(bbox_to_anchor=(1.0, 1.0), loc='upper left')
plt.savefig('computed flowrate 2019.png', bbox_inches='tight') 
plt.show() 
```
